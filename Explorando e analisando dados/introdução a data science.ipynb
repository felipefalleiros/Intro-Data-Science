# Melhorando nossa visualiaÃ§Ã£o


```python
# Separando variÃ¡veis com quantidade de valores de filmes com a linga en e o geral
total_por_lingua = tmdb['original_language'].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
# Cria a variavel com a quantidade de linguas diferentes de en
total_do_resto = total_geral - total_de_ingles
print(total_geral, total_de_ingles, total_do_resto)
```

```python
# Criando um novo df
# A estrutura para criar um df Ã© a mesma de um dicionÃ¡rio em python
dados =  {
  'lingua': ['ingles', 'outros'],
  'total': [total_de_ingles, total_do_resto]
    }
dados = pd.DataFrame(dados)
dados
```

```python
# VisualizaÃ§Ã£o de um grÃ¡fico de barras do novo df
sns.barplot(data=dados, x='lingua', y='total')
```

```python
# VisualizaÃ§Ã£o de um grÃ¡fico de piza do df criado
dados.plot(kind='pie', labels=dados['lingua'], y='total')
```

```python
# Realizando a contagem dos valores diferentes de en
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_de_outros_filmes_por_lingua.head()
```

```python
# plotando o grÃ¡fico da informaÃ§Ã£o acima
sns.countplot(data=tmdb.query('original_language != "en"'), x='original_language')
```

```python
# Fazendo a ordenaÃ§Ã£o, alteraÃ§Ã£o de cor e tamanho
plt.figure(figsize=(16,8))
sns.countplot(data=tmdb.query('original_language != "en"'),
              order=total_de_outros_filmes_por_lingua.index,
              hue='original_language',
              hue_order=total_de_outros_filmes_por_lingua.index,
              palette='mako',
              stat='percent',

              x='original_language',)
plt.title('Quantidade de filmes por lingua exceto inglÃªs')
plt.show()
```
