Uma empresa que presta serviços de consultoria e aconselhamento para pessoas que querem migrar do Brasil para o Canadá deseja melhorar os seus serviços. Para isso, temos alguns arquivos que foram disponibilizados para fazer uma análise das tendências de imigração do Brasil dos últimos anos.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
```

```python
df = pd.read_csv(url)
df
```

```python
df.info()
```

Vamos começar com a primeira tarefa desse projeto que é: **analisar as tendências de imigração do Brasil ao longo dos anos.**

```python
# Alterando o index do df para País
df.set_index('País', inplace=True)
```

```python
# Criando uma lista com os anos do df transformando os anos em str
anos = list(map(str, range(1980, 2014)))
anos
```

```python
# Salvando na variavel brasil o intervalo que pega a linha de index Brasil e as colunas dentro da variavel anos
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
# Criando um dict com a chave anos com valores sendo os indices da series brasil
# e os valores sendo os valores da series brasil
# to_list() transforma a series em uma lista
brasil_dict = {'anos':brasil.index.to_list(),
               'imigrantes':brasil.values.tolist()
               }
# Transformando o dict em df
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
# Alterando o tamanho do gráfico (largura, altura)
plt.figure(figsize=(8, 4))
# plotando um gráfico de linhas com os anos no eixo X e a quantidade de imigrantes no eixo Y
plt.plot(dados_brasil['anos'], dados_brasil['imigrantes']);
# Adicionando um título para o gráfico
plt.title('Imigração do Brasil para o Canadá')
# Adicionando um rótulo para as informações do eixo X e Y
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
# Definindo os anos que irão aparecer no eixo X
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']);
```

```python
# fig para criar uma figura que seria quadro em branco
# ax representa os eixos pois podemos ter mais de um gráfico dentro de uma mesma figura
fig, ax = plt.subplots(figsize=(8,4))

# setando os eixos X e Y
ax.plot(dados_brasil['anos'], dados_brasil['imigrantes'])

# setando o titulo com uma quebra de linha \n
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')

# setando os rotúlos para cada eixo
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')

# Modificando o eixo X e setando um intervalor de 5 em 5 anos
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
# 1 representa as linhas dentro da figura e 2 as colunas
fig, axs = plt.subplots(1,2, figsize=(15,5))

# axs[0] representa o gráfico da primeira coluna de índice 0
axs[0].plot(dados_brasil['anos'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

# adiciona uma grade atrás do gráfico pra melhor visualização dos valores
axs[0].grid()

# axs[1] representa o gráfico da segunda coluna de índice 1
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')

# adiciona uma grade atrás do gráfico pra melhor visualização dos valores
axs[1].grid()

plt.show()
```

# Comparando tendências de imigração Brasil/Argentina para o Canadá

```python
# Selecionando os index Brasil e Argentina e a lista de colunas anos
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao
```

```python
# Fazendo a transposição transformando as linhas em colunas e vice versa
df_comparacao = df_comparacao.T
df_comparacao
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil', color='Blue')
plt.plot(df_comparacao['Argentina'],label ='Argentina', color='Red')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))

# Alterando o espaçamento vertical (hspace) e horizontal (wspace)
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Adicionando um titúlo geral na figura
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

# Aplicando alterações em todos os eixos ao mesmo tempo utilizando o atributo flat
# que é um iterador permite percorrer um array MultD como se fosse 1D
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')


ymin = 0
ymax = 7000

# ravel retorna uma cópia do array em 1 dimensão
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)


plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\n para o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Removendo o fundo do gráfico
ax.set_frame_on(False)

# Removendo o eixo x (valores e titulo)
ax.get_xaxis().set_visible(False)


ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

# Gráfico com animação

```python
dados_brasil['ano'] = dados_brasil['anos'].astype(int)
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]],
               mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),

    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),

    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1],
 y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]

fig.frames = frames

# Mostrando a figura
fig.show()
```
